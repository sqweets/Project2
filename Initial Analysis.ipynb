```python
#
# Get the F/EF values from the decade dataframes counts
#
def formatDataForPlotting(decade_counts, decade):
    f_result = []
    ef_result = []
    
    if (decade == "00"):
        # Get both F annd EF data and add together
        f_result = f_scales(decade_counts)
        ef_result = ef_scales(decade_counts)
        
        # Add the F and EF scale values
        result = [sum(x) for x in zip(f_result, ef_result)]
    elif (decade == "10"):
        # Get EF only
        result = ef_scales(decade_counts)
    else:
        # Get F only
        result = f_scales(decade_counts)
    
    return result

#------------------------------------------
def f_scales(decade_counts):
    f_result = []

    try:
        f_result.append(decade_counts.F0)
    except AttributeError:
        f_result.append(0)

    try:
        f_result.append(decade_counts.F1)
    except AttributeError:
        f_result.append(0)

    try:
        f_result.append(decade_counts.F2)
    except AttributeError:
        f_result.append(0)

    try:
        f_result.append(decade_counts.F3)
    except AttributeError:
        f_result.append(0)

    try:
        f_result.append(decade_counts.F4)
    except AttributeError:
        f_result.append(0)

    try:
        f_result.append(decade_counts.F5)
    except AttributeError:
        f_result.append(0)
        
    return f_result

#------------------------------------------
def ef_scales(decade_counts):
    ef_result = []

    try:
        ef_result.append(decade_counts.EF0)
    except AttributeError:
        ef_result.append(0)

    try:
        ef_result.append(decade_counts.EF1)
    except AttributeError:
        ef_result.append(0)

    try:
        ef_result.append(decade_counts.EF2)
    except AttributeError:
        ef_result.append(0)

    try:
        ef_result.append(decade_counts.EF3)
    except AttributeError:
        ef_result.append(0)

    try:
        ef_result.append(decade_counts.EF4)
    except AttributeError:
        ef_result.append(0)

    try:
        ef_result.append(decade_counts.EF5)
    except AttributeError:
        ef_result.append(0)
        
    return ef_result
```
